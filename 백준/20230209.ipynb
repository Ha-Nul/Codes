```python
def Number(X):
    N=int(X)
    N_list=list(range(1,N+1))
    count=0

    for i in range(0,N):
        if N_list[i] < 10:
            count+=1
        if N_list[i] >= 10 and N_list[i] < 100:
            count+=1
        if N_list[i] > 100 and N_list[i] < 1000:
            A=N_list[i]//100
            B=(N_list[i]-A*100)//10
            C=N_list[i]-A*100-B*10

            if A-B == B-C:
                count+=1
            else:
                None
        elif N_list==1000:
            None
        
    print(count, end="")
```
