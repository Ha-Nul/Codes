```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

Low Gamma

```python
a0 = lambda g: -(1/2)*g**2 + (7/128)*g**4 - (29/2304)*g**6 + (68687/18874368)*g**8
b1 = lambda g: 1 - g - (1/8)*g**2 +(1/64)*g**3 - (1/1536)*g**4 - (11/36864)*g**5 + (49/589824)*g**6 - (55/9437184)*g**7 - (265/113246208)*g**8
```

```python
gamma = 0.03
```

```python
print(b1(gamma)-a0(gamma))
```

```python
a0(gamma)
```

```python
b1(gamma)
```

High Gamma

```python
ha0 = lambda g: 2*np.sqrt(g) - 2*g
hb1 = lambda g: 16*np.sqrt(2/np.pi)*g**(0.75) * np.exp(-4*np.sqrt(g)) + 2*np.sqrt(g) - 2*g
```

```python
hb1_arr = [hb1(i) for i in range(10)]
ha0_arr = [ha0(j) for j in range(10)]
gam_arr = [k for k in range(10)]
```

```python
plt.grid()

plt.xlim([0,9])
plt.plot(gam_arr,hb1_arr)
plt.plot(gam_arr,ha0_arr)
```
