```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_expression(x, g):
    return (np.sin(x/2) - 
            (g * np.sin(3*x/2)) / 16 + 
            (g**2 * (np.sin(3*x/2) + np.sin(5*x/2))) / 256 - 
            (g**3 * (np.sin(3*x/2)/3 + 4*np.sin(5*x/2)/9 + np.sin(7*x/2)/18)) / 4096 + 
            (g**4 * (-11*np.sin(3*x/2)/9 + np.sin(5*x/2)/6 + np.sin(7*x/2)/12 + np.sin(9*x/2)/180)) / 65536)
```

```python
x_arr=np.linspace(0,4*np.pi,101)
xarr = np.zeros(len(x_arr))

for i in range(len(x_arr)):
    xarr[i] = calculate_expression(x_arr[i],2)
```

```python
plt.plot(x_arr,xarr)
```
