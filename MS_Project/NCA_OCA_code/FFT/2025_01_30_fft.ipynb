### FFT function

```python
def FFT(location):
    data = pd.read_csv(location,header=None,delim_whitespace=True)
    
    tau = data[0]
    freq = np.fft.fftfreq(len(tau), tau[1] - tau[0])
    transformed_data = np.fft.fft(data[1])

    return [freq, transformed_data]
```
