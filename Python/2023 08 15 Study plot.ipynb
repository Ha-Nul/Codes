```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ln(g,D,x):
    a = np.abs((D+x)/(D-x))
    return -g*np.log(a)
```

```python
def A(g,omega,U,D,ep):
    func = g/((omega-ep+U-ln(g,D,omega))**2+(g/2)**2)

    return func
```

```python
omega = np.linspace(-23,23,100)

D =[]
E =[]

for i in range(len(omega)):
    B = A(5,omega[i],1,8,3)
    C = A(20,omega[i],1,20,3)

    D.append(B)
    E.append(C)
```

```python
F = np.array(D)
G = np.array(E)

plt.plot(omega,D)
plt.plot(omega,E)
plt.show()
```

```python
def x_y(x,y):

    def n_up(n):
        return np.arctan(n-x)*y

    def n_down(n):
        return np.arctan(n-x)*y

    n_updown = n_up(n_down())-n_down(n_up())

    return n_updown
```
