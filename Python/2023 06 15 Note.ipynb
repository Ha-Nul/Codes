```python
def delta_2_cosh(x):
    import numpy as np
    delta = 1
    
    ch = np.exp((-delta/2)*(1-2*x))+np.exp((delta/2)*(1-2*x))
    #ch_n = np.linalg.norm(ch)

    #func=ch/ch_n

    return ch
```

```python
def tan(x):
    return 1-x*np.tanh(0.5)
```
