```python
import numpy as np
import sympy as sp
import scipy.constants as ct
import matplotlib.pyplot as plt

sp.init_printing(use_latex=True)
```

```python
def g_disp(l,c,W,E,L):
    m = sp.symbols('m')
    #m = np.linspace(0,10,100)

    alpha = ct.Planck / ( 4 * np.sqrt(l/c) * ct.elementary_charge**2)
    v = 1 / np.sqrt(l*c)
    nu = (ct.pi * ct.hbar * W) / (alpha * E)
    w = m * v * ct.pi * L   #omega

    g_first = 2 * ct.pi * v / (alpha * L)
    g_second = w / ( 1 + ( nu * w / W )**2)

    g = sp.sqrt(g_first*g_second)
    #g = np.sqrt(g_first * g_second)

    sp.plot(g,(m,0,10))
    #plt.plot(m,g)
    #plt.show()
    print(g)
```
