```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as cn
```

```python
def Function(t):
    n = 10
    return (1/(+0.01*t))*np.sin(n*t)
```

```python
jmax = np.linspace(0,1,100)

plt.plot(jmax,Function(jmax))
plt.show()
```

```python
for i in range(10):
    plt.plot(jmax,Function(jmax-jmax[i+1]))
    plt.ylim(-1000,1000)
    plt.xlim(0,20)
    plt.show()
```
