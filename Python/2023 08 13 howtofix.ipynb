```python
def standard_even(r,z):

    A = [[0 for j in range(z)] for k in range(z)]

    for i in range(z):
        for j in range(z):
            try:
                if i==j:
                    A[i][j] = (i)**2
                if abs(i-j) == 1:
                    A[i][j] = -r/2
                else:
                    None
            except:
                None

    test_A = np.array(A)
    test_A[0][1] = -r/(np.sqrt(2))
    test_A[1][0] = -r/(np.sqrt(2))

    return test_A
```

```python
def eigenvec_even(r,z,m):

    Matrix = standard_even(r,z)
    A_eig = np.linalg.eig(Matrix)
    A_trans = np.transpose(A_eig[1])

    return A_trans[m]
```

```python
def elements(d,x,y):

    N = standard_even(0,d)

    arr = np.matmul(N,y)
    arr2 = np.matmul(x,arr)

    return arr2
```

```python
def Mat(r,d):
    Mat = []

    for i in r:
        one = eigenvec_even(i,d,0)
        two = eigenvec_even(i,d,1)

        rix = [[elements(d,one,one),elements(d,one,two)],[elements(d,two,one),elements(d,two,two)]]
        Mat.append(rix)

        #print(rix)

    return Mat
```
