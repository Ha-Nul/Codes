```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def self_Energy(coupling,omega,ek,mu,eta):
    if type(coupling) != int:
        coup = np.zeros(len(coupling))
        for i in range(len(coupling)):
            func = ((coupling[i])**2)/(omega[i]-ek[i]+mu+eta*(1j))
            coup[i] = np.real(func)+np.imag(func)
        coup_sum = np.sum(func)

        print(coup_sum)
        return coup_sum
    
    if type(coupling) == int:
        func = coupling/(omega-ek+mu+eta*(1j))
        return func



def A(coupling,omega,ed,mu,ek,U,eta):
    func = 1/(omega-ed+mu-U-self_Energy(coupling,omega,ek,mu,eta))
    return -2*np.imag(func)

def B(omega,ed,mu,U,self_s):
    func = 1/(omega-ed+mu-U-self_s)
    return -2*np.imag(func)
```

```python
omega=np.linspace(-10,10,200)
coupling=np.linspace(0,0.1,100)
ek = np.linspace(0,1,100)

func = B(omega,4,3,1,self_Energy(coupling,omega,ek,3,0.01))
```
