```python
import numpy as np
import scipy.constants as ct
import matplotlib.pyplot as plt

#create simple diffrential stuff

x=np.linspace(-2,2,50)
y=x**2

print(y[:50])
y_value=np.zeros(50)
```

```python
dx = 4/200

for i in range(50):
    y_value[i+1] = (((y[i]+dx) - (y[i]-dx))/dx)*x[i]

    print(y_value)

    if i+2 >= 50:
        break

plt.plot(x,y)
plt.plot(x,y_value)
plt.show()
```
