```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as ct
from scipy import signal

c0 = ct.speed_of_light
lambda_min = 350e-9

dx = lambda_min/20
dt = dx/c0
```

```python
def Source_Function(t):
    lambda_0 = 550e-9
    w0 = 2*np.pi*c0/lambda_0
    tau = 300
    t0 = tau*3

    return np.exp(-(t-t0)**2/tau**2)*np.sin(w0*t*dt)
```

```python
t=np.linspace(-10,2000,100000)

plt.plot(t,Source_Function(t))
```

```python
f_trans=np.fft.fft(Source_Function(t))
```

```python
plt.plot(t,f_trans)
```

```python
plt.plot(Source_Function(t))
plt.show()
```

```python
signal.unit_impulse(8)

imp = signal.unit_impulse(100, 'mid')
b, a = signal.butter(4, 0.2)
response = signal.lfilter(b, a, imp)
```

```python
plt.plot(np.arange(-50, 50), imp)
plt.plot(np.arange(-50,50), response)
plt.margins(0.1, 0.1)
plt.xlabel('Time [samples]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
