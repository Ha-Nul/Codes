```python
import numpy as np
import matplotlib.pyplot as plt
import random
import scipy.constants as ct
```

```python
i = random.randint(0,9)
theta = np.linspace(0,1,10)

h = ct.h
e = ct.elementary_charge
C = 2*np.pi*theta[i]

flux = h/(2*e)
I = np.pi * 4.4/100
```

```python
E = I*flux/(2*np.pi)
V = 2*e/C
```

```python
H_init = np.zeros(10)

for j in range(9):
    H_init[j] = V**2 - E*np.cos(theta[j] - theta[j+1])
    

H = H_init * C/2
```

```python
x_pos = np.arange(0,10)
y_pos = np.zeros(10)

x_direct = np.zeros(10)
y_direct = np.zeros(10)

for i in range(10):
    x_direct[i] = np.cos(theta[i])
    y_direct[i] = np.sin(theta[i])
```

```python
fig, ax = plt.subplots(figsize = (12, 7))
ax.quiver(x_pos, y_pos, x_direct, y_direct,
         scale = 15)

plt.show()

print(np.sum(H))
```
