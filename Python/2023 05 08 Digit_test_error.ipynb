```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

sp.init_printing()
```

```python
def Q2_error(n,m):
    x=np.linspace(0,10,100)
    
    Q2_Paper = 0.0002*x**6 - 0.0067*x**5 + 0.0815*x**4 - 0.5171*x**3 + 1.9482*x**2 - 4.7087*x + 9.8855
    Q2_mc = 0.0003*x**6 - 0.0076*x**5 + 0.0927*x**4 - 0.5845*x**3 + 2.1503*x**2 - 4.9792*x + 9.9652

    n_paper = Q2_Paper[n:m]
    n_mc = Q2_mc[n:m]

    Q2_error = np.abs((n_paper - n_mc)*100/(n_paper))
    Q2_avg = np.average(Q2_error)

    plt.plot(x[n:m],Q2_error)

    print((np.ceil(Q2_error*100)/100))
    print(Q2_avg)
```

```python
Q2_error(0,100)
```
