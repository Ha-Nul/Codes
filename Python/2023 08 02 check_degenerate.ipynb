```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def standard_even(r,z):

    A = [[0 for j in range(z)] for k in range(z)]

    for i in range(z):
        for j in range(z):
            try:
                if i==j:
                    A[i][j] = (i)**2
                if abs(i-j) == 1:
                    A[i][j] = -r/2
                else:
                    None
            except:
                None

    test_A = np.array(A)
    test_A[0][1] = -r/(2)
    test_A[1][0] = -r/(np.sqrt(2))
    A_eig=np.linalg.eig(test_A)

    #3print(test_A)

    return A_eig[0]
```

```python
standard_even(1,10)
```

```python
def standard_odd(r,z):

    A = [[0 for j in range(z)] for k in range(z)]

    for i in range(z):
        for j in range(z):
            try:
                if i==j:
                    A[i][j] = (i+1)**2
                if abs(i-j) == 1:
                    A[i][j] = -r/2
                else:
                    None
            except:
                None

    test_A = np.array(A)
    A_eig=np.linalg.eig(test_A)

    #print(A_eig[0])

    return A_eig[0]
```

```python
r = np.linspace(-10,10,500)
arr = []
arr_e = []

for i in r:
    arr.append(standard_odd(i,9))
    arr_e.append(standard_even(i,10))

app = np.sqrt((24*r**4)/400)
fig, ax = plt.subplots(figsize=(10,8))

Energy1 = 1 - (1*r**2)/12
Energy2 = 1 + (5*r**2)/12

#Energy1 = 1 + np.zeros(500)
#Energy2 = 1 + r**2/2

#plt.plot(j,np.zeros(500))
ax.plot([1])
spines = ax.spines
[i.set_linewidth(2) for i in spines.values()]
plt.tick_params(axis='x',direction='in')
plt.tick_params(axis='y',direction='in')

plt.ylim(-2,14)
plt.ylabel('Eigenvalues(odd)')
plt.xlabel(r'$\gamma$')


plt.plot(r,arr)
plt.plot(r,arr_e)

plt.plot(r,Energy1,label='sp_Eigen_high',color='limegreen')
plt.plot(r,Energy2,label='sp_Eigen_low',color='red')
plt.legend(loc='upper right')
plt.grid(True,linestyle='--')
plt.show()
```
