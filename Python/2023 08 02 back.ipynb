```python
a = int(input())

b=1

for i in range(1,a+1):
    b=b*i

c = list(map(int,str(b)))
d = c[::-1]

k = 0

for j in d:
    if j!=0 and len(d) != 1:
        break
    elif len(d) == 1:
        print(0)
    else:
        k+=1

if len(d) != 1:
    print(k)
```
