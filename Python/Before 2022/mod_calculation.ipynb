```python
from matplotlib import pyplot as plt
```

```python
def Mod_Leftover_bi(comp1,comp2):
    
    print('지수함수의 밑이',comp1,',',comp2,'인 두 경우의 모듈러 연산 계산 결과','\n')
    
    N=int(input("N의 값= "))
    x=int(input("x의 값= "))
    
    c1=int(comp1)
    c2=int(comp2)
    
    x_list=[]
    c1_list=[]
    c2_list=[]

    for i in range(0,x,1):
        
        x_list.append(i)
        
        f1=(c1**i)%N
        c1_list.append(f1)
        
        f2=(c2**i)%N
        c2_list.append(f2)
            
    print(x_list,c1_list,c2_list, sep='\n')
    #print(c1_list.index(1),c2_list.index(1))
    
    plt.plot(c1_list,'--',color='olivedrab')
    plt.plot(c2_list,'--',color='dodgerblue',marker='o')
    plt.xlabel('Count')
    plt.ylabel('x mod N')
    plt.title('Mod result in two number')
    plt.show()
    
    return x_list,c1_list,c2_list
```

```python
Mod_Leftover_bi(7,135)
```
