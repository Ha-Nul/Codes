```python
import numpy as np
import math
import sympy as sp
from sympy.plotting import plot

sp.init_printing(use_latex=True)
```

```python
a=sp.symbols('a')
y=a+1

plot(y,(a,-5,5))
```

```python
x, y, z=sp.symbols('x,y,z')
```

```python
test=sp.cos((math.pi/2)*x+1)
test.sp.sub(x,0)

plot(test)
```

```python
expr=sp.sin(2*x)+sp.cos(2*x)
expr_E=sp.expand_trig(expr)

plot(expr_E)
```
