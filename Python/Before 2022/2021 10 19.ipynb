```python
import sympy as sp
import math
from sympy.plotting import plot

sp.init_printing(use_latex=True)
```

```python
x=sp.symbols('x')
#a,b=sp.symbols('a b')
a=10
b-20
f=sp.symbols('f',cls=sp.Function)
v=f(x)
```

```python
d_eq=sp.Eq(sp.diff(f,x),-a*v-b*(x**2)*v)
```

```python
sol=sp.dsolve(d_eq,v)
```

```python
sol
```

```python
f_x=-1/(10*((2*x**2)+1))
plot(f_x)
```

```python
plot(sol.rhs)
```
