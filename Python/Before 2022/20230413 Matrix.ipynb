```python
from sympy import *
import numpy as np

init_printing(use_latex=True)
```

```python
C_J , C_P , C_x = symbols('C_J , C_P , C_x')
```

```python
# For N=4

A_bar = Matrix([[1/C_P,0,0,0],[0,1/C_x,0,0],[0,0,1/C_x,0],[0,0,0,1/C_x]])
B_bar = Matrix([[1,-1,0,0],[-1,2,-1,0],[0,-1,2,-1],[0,0,-1,2]])
```

```python
B_bar
```

```python
# Diagonalize A

A_ec=A_bar.eigenvects()
A_ev=A_bar.eigenvals()

#A_pr=dict(zip(A_ev.values(),A_ev.keys()))
A_arr=list(A_ev.keys())
A_pwarr=[]

for i in range(4):
    if i < 2:
        A_pwarr.append(sqrt(A_arr[i]))
    else:
        A_pwarr.append(sqrt(A_arr[1]))

A_pw=Matrix([[A_pwarr[0],0,0,0],[0,A_pwarr[1],0,0,],[0,0,A_pwarr[2],0],[0,0,0,A_pwarr[3]]])
```

```python
# A의 Eigenvector가 이미 orthogonalized된 형태이므로 바로 곱셈함.
S11=A_pw*B_bar*A_pw
```

```python
S12=S11.eigenvects()
```
