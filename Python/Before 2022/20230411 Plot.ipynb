```python
import numpy as np
import sympy as sp
from sympy.plotting import plot
import matplotlib.pyplot as plt

sp.init_printing(use_latex=True)
```

```python
def p1(a1,b1,c):
    val=((-b1+(b1**2 - 4*a1*c)**0.5)/2*c)**0.5
    return val

def p3(b2,c):
    val2=(-b2/(2*c))**0.5
    return val2

x=sp.symbols('x')

def reg1(a,b,c,h):
    int=sp.integrate((x*(x-p1(a,b,c))*(x+p1(a,b,c))-h),x)
    return int

def reg2(c,h):
    int=sp.integrate(x-h,x)
    return int

def reg3(b,c,h):
    int=sp.integrate((x*((x-p3(b,c))**2)*((x+p3(b,c))**2)-h),x)
    return int


r1=reg1(-1,1.9,1,0)
r2=reg2(0,100)
r3=reg3(0,1,100)
```

```python
from sympy.calculus.util import *
```

```python
ivl = Interval(-10,10)
print(minimum(r1, x, ivl))
print(maximum(r1, x, ivl))
print(stationary_points(r1, x, ivl))
```

```python
ivl = Interval(-200,200)
print(minimum(r2, x, ivl))
print(maximum(r2, x, ivl))
print(stationary_points(r2, x, ivl))
```

```python
ivl = Interval(-5,5)
print(minimum(r3, x, ivl))
print(maximum(r3, x, ivl))
print(stationary_points(r3, x, ivl))
```
