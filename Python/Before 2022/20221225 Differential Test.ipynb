Do Differential in high-school method

```python
def f(x):
    return x**2 + x
```

```python
derivative(f,1)
```

```python
cal_space=np.linspace(0,5,100)
blank_space=np.zeros(100)
fun_space=np.zeros(100)

for i in range(0,99,1):
    cal_point=cal_space[i]
    cal_result=derivative(f,cal_point)
    fun_result=f(cal_point) #Function class doesn't need to change parameter datatype for use.

    blank_space[i]=cal_result
    fun_space[i]=fun_result

def f_dev(x):
    return 2*x 

plt.plot(cal_space,blank_space)
plt.plot(cal_space,fun_space)
plt.show()

```

```python
test=f_dev(cal_space)

plt.plot(cal_space,test)
```

수치미분법을 이용한 접선그리기

```python
def numerical_diff(f,x):
    h=1e-10
    return (f(x+h)-f(x-h)/2*h)

def tangent_line(f,x):
    d=numerical_diff(f,x)
    y=f(x)-d*x

    #이거 리턴값에 들어간 코드가 뭔 뜻이냐
    return lambda t:d*t + y
```

```python
y=f(x)

tf=tangent_line(f,5)
y2=tf(x)

plt.plot(x,y)
plt.plot(x,y2)
```
