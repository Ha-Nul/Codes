```python
import sympy as sp
from sympy import *
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=4,3

x=sp.symbols('x')
Eq=2*x*(x**2-4*x+2)
Eq_2=(x**2)

plot(Eq,Eq_2,(x,0,4.5))
#solve(Eq(Eq,Eq_2),x)
solve(Eq(2*x*(x**2-4*x+2),x**2),x)
```

```python
integrate(abs(Eq-Eq_2),(x,0,4))
```

```python
def hash2bit(variable, hash_type):
    if hash_type =='V':
        bit=(variable<0.5)
    elif hash_type == 'H':
        bit=(variable<0.25)

    bit=str(int(bit))

    return bit
```
