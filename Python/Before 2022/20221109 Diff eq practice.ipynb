```python
import sympy as sp
sp.init_printing(use_latex=True)
```

```python
x=sp.symbols('x')
f=sp.symbols('f',cls=sp.Function)
y=f(x)

Diff_equation1=sp.Eq(sp.diff(y,x)*x**2-x*y,1/x)
```

```python
sp.dsolve(Diff_equation1)
```

```python
Diff_equation2=sp.Eq(x*sp.log(y)*sp.diff(y,x)-y*sp.log(x),0)
```

```python
sp.dsolve(Diff_equation2)
```

```python
Diff_equation3=sp.Eq(sp.diff(y,x,x,x)+2*sp.diff(y,x,x)+2*sp.diff(y,x),0)
```

```python
sp.dsolve(Diff_equation3)
```

```python
Diff_equation5=sp.Eq(2*x-y*sp.sin(2*x),(sp.sin(x)**2-2*y)*sp.diff(y,x))
```

```python
sp.dsolve(Diff_equation5)
```
