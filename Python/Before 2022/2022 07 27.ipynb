```python
wavelength=15.0 #파장이 15cm인 물결
k=np.pi*2/wavelength #파장이 주어졌을 때의 파수

side=100 #한쪽 변이 100cm라고 가정
points=500 #한 변을 500개의 점으로 나누어 계산
spacing = side/points #100cm를 500개로 나누었을 때의 간격

A=np.empty([points,points],float) #데이터를 기록할 빈 행렬
for i in range(points):
    y=spacing*i
    for j in range(points):
        x=spacing*j
        r=np.sqrt(x**2+y**2) #(0,0)에서의 거리를 계산
        A[i,j]=np.sin(k*r) #원형파형을 만듬
        
plt.imshow(A,origin="lower",extent=[0,side,0,side]) #(0,0)인 점을 아래에 그린다
plt.xlabel("x(cm)")
plt.ylabel("y(cm)")
plt.colorbar()
plt.jet()#gray() #흑백으로 그래프를 표시. 선택하지 않으면 기본값은 viridis()
plt.show()
```

```python
N=10
sum=0.0

for n in range(N):
    x=(-1)**n/(2*n+1)
    sum+=x
    
    print(n,4*sum)
```

```python
N=10000
eps=1E-4
sum=0.0
n=0
while(n<N):
    x=(-1)**n/(2*n+1)
    sum+=x
    
    if(abs(x)<eps):break
    if(n%100==0):
        print(n,4*sum)
    n+=1
print(n,abs(x),4*sum,np.pi)
```

```python
eps=1E-8 #얻고자 하는 해의 유효숫자
def f(x):
    return np.sin(x)-0.5 #f(x)를 sin(x)-1/2로 정의

x1=0; x2=np.pi/2.0
while((x2-x1)>eps):
    x3=(x1+x2)/2.0
    
    if (f(x1)*f(x3)<0.0):x2=x3
    else:
        x1=x3
    print(x3,f(x3))
```
