```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc

import sympy as sp
```

```python
def simple_plot_1(l,c,x,L):
    mode = pd.read_csv('2023-05-01 test_csv.CSV',header=None,names=['No','mod'])
    m = np.array(mode['mod'])
    k = m*(np.pi)/L
    v = 1 / np.sqrt(l*c)

    omega = (2*v/x)*np.sin(k)

    plt.plot(k)
```

```python
simple_plot_1(1,1,5)
```

```python
def simple_plot_2(W):
    mode = pd.read_csv('2023-05-01 test_csv.CSV',header=None,names=['No','mod'])
    m = np.array(mode['mod'])
    
    #m=np.linspace(0,50,100)
    M=(len(m)-1)
    k = m*(np.pi)/(2*M)

    omega = W*np.sin(k)

    plt.plot(k,omega)
    plt.xlim([0,2])
    plt.show()               
```

```python
simple_plot_2(10)
```
