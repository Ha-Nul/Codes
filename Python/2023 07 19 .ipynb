```python
import numpy as np
from scipy import special as sp
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def Eipos(x,t):
    one = sp.expi(-t*x)-np.exp(t*x)/t
    return one
    
def Eineg(x,t):
    one = sp.expi(t*x)
    return one 
```

```python
omega = np.linspace(-10,10,500)
tau = np.linspace(-1,1,500)
tau1 = np.linspace(-1,0,500)
tau2 = np.linspace(0,1,500)

X,Y = np.meshgrid(omega,tau1)
N,M = np.meshgrid(omega,tau2)
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,Eineg(X,Y),cmap='cool')
ax.plot_surface(N,M,Eipos(N,M))
ax.view_init(30,60)

plt.show()
```
