```python
import numpy as np
from scipy import special as sp
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def hyp(a,b,c,z):
    mid = a*b*z/c
    fin = a*(a+1)*b*(b+1)*z**2/(2*c*c+1)

    return 1+mid+fin
```

```python
omega = np.linspace(-10,10,500)
tau = np.linspace(0,1,500)

X,Y = np.meshgrid(omega,tau)

def function(x,y):
    return -np.exp(x*y)*hyp(1,y+1,y+1,np.exp(x))/y

def function2(x,y):
    return -np.exp(x*y)*sp.hyp2f1(1,y+1,y+1,np.exp(x))/y

def function3(x,y):
    return -np.exp((x+1)*y)*sp.hyp2f1(1,y+1,y+2,np.exp(x))/(y+1)

def over(t):
    return 1/t
```

```python
plt.figure(figsize=(10,8))
plt.plot(tau,function3(-10,tau)-function3(-1.018,tau),label='Test function')
plt.plot(tau,over(tau),label=r'$1/\tau$')

plt.xscale('log')
plt.yscale('log')

plt.xlabel(r'$\tau$(log scale)')
plt.ylabel(r'Value (log scale)')

plt.tick_params(axis='x',direction='in')
plt.tick_params(axis='y',direction='in')

plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,function3(X,Y),label='curve')
ax.view_init(-10,60)

ax.set_xlabel(r'$omega$')
ax.set_ylabel(r'$\tau$')
ax.set_zlabel(r'Function Value')

plt.show()
```
